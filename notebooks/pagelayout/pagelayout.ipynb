# Bacis components of Page Layout

```python
from grizly import Text, Row, Column, Page, GridCardItem, GridLayout, Layout
```

### Class Page

```python
# It is a main container which wraps content into html template
print(Page(content='').to_html())
# Saving html to file
Page(content='').write_out(path='pagelayout.txt')
```

### Class Text

```python
Page([Text("header", style="color:red"),Text("footer", style="color:red")])
```

### Class Column & Row

```python
Page([Row([Column([Text("row1column1")]), 
           Column([Text("row2column1"),
                   Text("row2column2")]),
           # columns are placed automatically to fullfill all space
           Column([Text("row3column1"), 
                   Text("row3column2"),
                   Text("row3column3")]),
           # we can set up column width passing one argument for all or set it individualy for each column. 
           Column([Text("row4column1"), 
                   Text("row4column2"),
                   Text("row4column3")],size=[1]),
           Column([Text("row5column1"), 
                   Text("row5column2"),
                   Text("row5column3")],size=[1,6,4]),
           # box has size of 12 units, so sum of the sizes needs to be less than that.
           Column([Text("row6column1"), 
                   Text("row6column2")],size=[6,7]),
           # empty row don't work
           Column([Text(" ")]),
           # without the Column() text will be placed badly 
           Text("row7")])])
```

```python
# we need to have a row before column
Page([Column([Text('col1'),Text('col2')])])
```

### Layout

```python
# with layout we can nest rows and columns ()
body=Row([
    Column([
        Row([
            Column([
                Text('col1row1.1')]),
            Column([
                Text('col1row1.2')])]),
        Text('col2row1')])])
Page([Layout(body=body)])
```

### GridLayout

```python
g_example = GridCardItem(header='Header',paragraph='Paragraph',href='link',img='grid_img.png')
g_empty = GridCardItem()
header1 = "Layout header1"
body1 = Row([Column([g_example,
                    g_example,
                    g_example,
                    g_example]),
            Column([g_example,
                    g_example,
                    g_example,
                    g_empty])])
header2 = "Layout header2"
body2 = Row([Column([g_example,
                     g_example,
                     g_example],size=[2])])
 
Page([GridLayout(header1,body1),GridLayout(header2,body2)])
```
